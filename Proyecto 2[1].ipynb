# 2. Marco teórcico
### Regresión logística
Modela la probabilidad de que una observación pertenezca a una categoría particular. Es un método de clasificación, por lo que se usan regresiones para la predicción de una respuesta cualitativa. Este tipo de regresión utiliza la función logística para tomar una combinación lineal de las variables predictoras y transformarlas para que el resultado sea siempre un valor entre 0 y 1. 

### Máquinas de vectores de soporte
Modelos de aprendizaje para analizar datos, donde buscamos alejarnos lo más posible de la incertidumbre. Se busca tener el hiperplano  que mejor separa las clases en el espacio de características. El "mejor" hiperplano es aquel donde la "calle" es lo más ancha posible entre él y los puntos de datos más cercanos de cada clase. Para manejar datos que no son perfectamente separables, se utiliza un "margen suave", controlado por el hiperparámetro C. Este permite que algunas observaciones estén dentro del margen o incluso del lado incorrecto de la frontera, creando un equilibrio entre ajustar los datos y evitar el sobreajuste.

### Redes neuronales (Multi-layer perceptron)
Es un tipo de red neuronal que funciona como una generalización no lineal de la regresión logística. Es un modelo que aprende en etapas a través de sus capas:
- Capa de Entrada (Input Layer): Es la capa que recibe las variables predictoras iniciales. Estas son las "acciones" o "causas" que observa el modelo. Cada neurona de esta capa representa una característica de entrada.
- Capas Ocultas (Hidden Layers): Toman la información de la capa de entrada y aprenden a crear nuevas representaciones o características intermedias. Estas nuevas características son transformaciones no lineales de las entradas, y la red aprende qué combinaciones son las más útiles para la predicción.
- Capa de Salida (Output Layer): Toma la información procesada por las capas ocultas y produce el resultado final: la predicción o "reacción" del modelo. 
El poder del MLP radica en su capacidad para aprender automáticamente las complejas interacciones entre las "acciones" de entrada para predecir la "reacción" de salida.

### Kernels
Parte del principio de que las clases no pueden ser separadas por una simple línea o plano en su espacio original. Por lo tanto, cada punto de todas las características pasa por una función que cambia tu punto de vista del espacio. Se puede mover para ver si en algún punto se puede hacer la división

### Métricos para clasificación
Para evaluar el rendimiento de un modelo de clasificación, se utilizan varias métricas. 
- Exactitud (Accuracy): indica el porcentaje total de predicciones (tanto positivas como negativas) que el modelo acertó. 
- Precisión (Precision): mide la calidad de las predicciones positivas (de todos los que el modelo dijo que eran positivos, ¿cuántos lo eran realmente?).
-  Recall (Sensibilidad): mide la capacidad del modelo para encontrar todos los casos positivos (de todos los que realmente eran positivos, ¿cuántos encontró el modelo?).
- La Especificidad (Specificity) es la contraparte del Recall, midiendo qué tan bien el modelo identifica los casos negativos. Finalmente, el AUC (Área Bajo la Curva), derivada de la curva ROC*, es una métrica global que mide la capacidad del modelo para discriminar entre clases positivas y negativas en todos los umbrales de decisión, siendo 1.0 un clasificador perfecto y 0.5 uno aleatorio. 
*La Curva ROC (Receiver Operating Characteristic) es un gráfico que mide el rendimiento de un modelo de clasificación al mostrar qué tan bien puede discriminar entre clases, trazando la tasa de verdaderos positivos (Recall) contra la tasa de falsos positivos en todos los umbrales de decisión posibles.


### Hiperparámetros
Los hiperparámetros son las "configuraciones" de un modelo de aprendizaje que se establecen antes de que comience el proceso de entrenamiento. A diferencia de los parámetros, estos son aquellos que no se ajustan a beta. Cuando se realiza correctamente, el ajuste de hiperparámetros minimiza la función de pérdida, lo que significa que el rendimiento del modelo se entrena para que sea lo más preciso posible. 

### Optimización Bayesiana
Se basa en el teorema de bases, donde se actualizan las creencias a partir de nueva información. Partimos de una función objetivo desconocida, hacemos el proceso Gaussiano normal, buscamos un punto de partida, efectuamos una regresión, tabulamos el modelo con intervalos, buscamos el siguiente valor y evaluamos el valor de la nueva x en la caja negra

*Con la regresión gausiana, asumimos que hay una distribución normal multivariada y queremos que se parezca a los datos que tenemos. Es decir, ¿cuánto tiene que valer la media y la desviación estándar de la multivariada para que se parezca a la distribución normal
Es un método eficiente para encontrar el mejor conjunto de hiperparámetros para un modelo. Se basa en el teorema de bayes, donde se actualizan creencias a partir de nueva información. Parte de un modelo probabilístico, como el proceso gaussiano*, que estima cómo se comporta la métrica de rendimiento en función de los diferentes valores de hiperparámetros. Su principal función es decidir cuál es la siguiente combinación de hiperparámetros más prometedora para probar. Es así como, se va obteniendo nuva información para tener un conjunto de alto rendimiento.

## 6. Regresion Logistica

```python
model_lg = LogisticRegression(max_iter = 1000, random_state=42, C=1)
pipeline_lg = Pipeline([
    ('scaler', StandardScaler()),
    ('model_lg', model_lg)
])

pipeline_lg.fit(X, y)

scores_lg = cross_val_score(pipeline_lg, X, y, cv=10, scoring='roc_auc_ovr')
print("AUC promedio:", np.mean(scores_lg))
print("AUC desviación estándar:", np.std(scores_lg))
```

```python
search_spaces = [
    {
        'model_lg__penalty': Categorical(['l1']),
        'model_lg__C': Real(1e-4, 1e4, prior='log-uniform'),
        'model_lg__solver': Categorical(['liblinear', 'saga'])
    },
    {
        'model_lg__penalty': Categorical(['l2']),
        'model_lg__C': Real(1e-4, 1e4, prior='log-uniform'),
        'model_lg__solver': Categorical(['lbfgs', 'liblinear', 'saga'])
    }
]

opt_lg = BayesSearchCV(
    estimator=pipeline_lg,
    search_spaces=search_spaces,
    n_iter=100,  
    cv=10,
    scoring='roc_auc_ovr',
    n_jobs=-1,
    verbose=1,
    random_state=42
)

opt_lg.fit(X, y)

mean_auc = opt_lg.best_score_
best_index = opt_lg.best_index_
std_auc = opt_lg.cv_results_['std_test_score'][best_index]

print("\n--- Resultados de BayesSearchCV (Corregido) ---")
print(f"Mejor puntuación (AUC): {mean_auc:.4f}")
print(f"Desviación estándar (AUC): {std_auc:.4f}")
print("Mejores hiperparámetros:")
print(opt_lg.best_params_)
```

## 7. Máquina de vectores de soporte con kernel RBF

```python
svc = SVC(kernel='rbf', probability=True, random_state=42, C=1, degree=2)
pipeline_svc = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', svc)
])
pipeline_svc.fit(X, y)
scores_svc = cross_val_score(pipeline_svc, X, y, cv=10, scoring='roc_auc_ovr')
print("AUC promedio:", np.mean(scores_svc))
print("AUC desviación estándar:", np.std(scores_svc))
```

```python
search_spaces = {
    'svc__C': Real(1e-4, 1e+4, prior='log-uniform') 
}
cv_strategy = KFold(n_splits=10, shuffle=True, random_state=42)

opt = BayesSearchCV(
    estimator=pipeline_svc,
    search_spaces=search_spaces,
    n_iter=100,  # 50 iteraciones de búsqueda
    scoring='roc_auc',
    cv=cv_strategy,
    n_jobs=-1,
    random_state=42,
    verbose=1 # Muestra el progreso
)
opt.fit(X, y)
```

```python
print("Mejor valor de C:", opt.best_params_['svc__C'])
print("Mejor AUC obtenido:", opt.best_score_)
print(f"Desviación estándar del mejor AUC: {opt.cv_results_['std_test_score'][opt.best_index_]:.4f}")
```

 ## 8. Multi-layer perceptron


```python
hidden_layers=(15,6)

model_mlp=MLPClassifier(random_state=10,hidden_layer_sizes=hidden_layers,max_iter=10000)
pipeline_mlp = Pipeline([
    ('scaler', StandardScaler()),
    ('model_mlp', model_mlp)
])
pipeline_mlp.fit(X, y)
scores_mlp = cross_val_score(pipeline_mlp, X, y, cv=10, scoring='roc_auc_ovr')
print("AUC promedio:", np.mean(scores_mlp))
print("AUC desviación estándar:", np.std(scores_mlp))
```

```python
def fun_obj(hi):
    global X, y
    model_mlp=MLPClassifier(random_state=10,hidden_layer_sizes=hi,max_iter=10000)
    pipeline_mlp = Pipeline([
        ('scaler', StandardScaler()),
        ('model_mlp', model_mlp)
    ])
    pipeline_mlp.fit(X, y)
    scores_mlp = cross_val_score(pipeline_mlp, X, y, cv=10, scoring='roc_auc_ovr')
    mean_score=np.mean(scores_mlp)
    
    return mean_score
```

```python
h1=[10,10]
h2=[20,5]
h3=[30,15]
hi=np.array([h1,
            h2,
            h3])
score_og=[]
for i in hi:
    mean_score=fun_obj(i)
    score_og.append(mean_score)

kernel=RBF(length_scale=5)
gp = GaussianProcessRegressor(kernel=kernel,random_state=10)

gp.fit(hi,score_og)
```

```python
h1=np.array([[10,10],
            [20,5],
            [30,15]])
score=score_og
auc=[]
for i in range(30):
    gp.fit(h1,score)
    x1n=np.random.uniform(10, 30,1000).reshape(-1,1)
    x2n=np.random.uniform(5, 15, 1000).reshape(-1,1)
    Xn=np.hstack((x1n,x2n))
    score_pred, score_std = gp.predict(Xn, return_std=True)
    score_upper=score_pred + 1.96*score_std
    score_lower=score_pred - 1.96*score_std
    h1_nueva = list(Xn[np.argmax(score_upper)])
    h1_nueva_tuple = tuple(np.round(h1_nueva).astype(int))
    score_nueva = fun_obj(h1_nueva_tuple)
    auc.append(score_nueva)
    h1 = np.vstack([h1, h1_nueva_tuple])
    score = np.hstack([score, score_nueva])
```

```python
print(f'Mi AUC optima es {score[np.argmax(score)]}, con las capas {h1[np.argmax(score)]} y se econtro en la interacion {np.argmax(score)}')
```

### 9.2 ROC por modelo

```python

from sklearn.metrics import roc_curve, auc
def plot_roc_cv(model_pipeline, X_data, y_data, title):
    cv = KFold(n_splits=10, shuffle=True, random_state=42)
    
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)
    
    plt.figure(figsize=(4, 4))
    
    # Iterar sobre cada fold de la validación cruzada
    for i, (train, test) in enumerate(cv.split(X_data, y_data)):
        # Entrenar el modelo
        model_pipeline.fit(X_data.iloc[train], y_data[train])
        # Predecir probabilidades en el conjunto de test
        y_pred_proba = model_pipeline.predict_proba(X_data.iloc[test])[:, 1]
        
        # Calcular ROC
        fpr, tpr, _ = roc_curve(y_data[test], y_pred_proba)
        
        # Interpolar TPR para tener una base común (mean_fpr)
        tprs.append(np.interp(mean_fpr, fpr, tpr))
        tprs[-1][0] = 0.0 # Asegurar que la curva empieza en (0,0)
        
        # Calcular y guardar AUC
        roc_auc = auc(fpr, tpr)
        aucs.append(roc_auc)
    
    # Graficar la línea de azar
    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='red', label='Azar (AUC = 0.50)', alpha=.8)
    
    # Calcular y graficar la curva ROC media
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0 # Asegurar que la curva termina en (1,1)
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    
    plt.plot(mean_fpr, mean_tpr, color='blue',
             label=rf'AUC Media = {mean_auc:.4f} $\pm$ {std_auc:.3f}',
             lw=3, alpha=.8)
    
    # Configuraciones finales de la gráfica
    plt.xlabel('Tasa de Falsos Positivos (FPR)', fontsize=12)
    plt.ylabel('Tasa de Verdaderos Positivos (TPR)', fontsize=12)
    plt.title(f'Curva ROC - {title}', fontsize=12, fontweight='bold')
    plt.legend(loc="lower right", fontsize=10)
    plt.grid(linestyle='--', alpha=0.6)
    plt.tight_layout()
    
    plt.show()


# --- 3. Definición de Modelos --- 
pipeline_lg = Pipeline([
    ('scaler', StandardScaler()),
    ('model_lg', LogisticRegression(
        C=8.35041760063047, 
        penalty='l2', 
        solver='lbfgs', 
        max_iter=1000, 
        random_state=42
    ))
])

pipeline_svc = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(
        C=0.07085355690283351, 
        kernel='rbf', 
        probability=True, 
        random_state=42
    ))
])

best_arch_mlp = (10, 6)
pipeline_mlp = Pipeline([
    ('scaler', StandardScaler()),
    ('model_mlp', MLPClassifier(
        hidden_layer_sizes=best_arch_mlp, 
        max_iter=10000, 
        random_state=10
    ))
])

# --- 4. Generar las 3 Gráficas (MODIFICADO) ---
# Gráfica 1: Regresión Logística

plot_roc_cv(pipeline_lg, X, y, 'Regresión Logística (Optimizada)')

# Gráfica 2: SVM (Kernel RBF)
plot_roc_cv(pipeline_svc, X, y, 'SVM (Kernel RBF) (Optimizado)')

# Gráfica 3: MLP (Red Neuronal)
plot_roc_cv(pipeline_mlp, X, y, f'MLP (Arq. Optimizada: {best_arch_mlp})')
```
